```python
from tkinter import *
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

train = pd.read_csv('C://CAPSTONE//DataSets//Heart//CSV//HeartTrain.csv')
#preping test labels and features
train_labels = train['Result']
train_features = train.drop('Result',1)
# Instantiate model with 5000 decision trees
rf = RandomForestRegressor(n_estimators = 5000, random_state = 100 , max_features = 0.25)
# Train the model on training data
rf.fit(train_features, train_labels)

def calc():
    a = age_var.get()
    b = gender_var.get()
    c = cpt_var.get()
    d = rbp_var.get()
    e = sc_var.get()
    f = fbs_var.get()
    g = rer_var.get()
    h = mhr_var.get()
    i = eia_var.get()
    j = ops_var.get()
    k = sop_var.get()
    l = nmv_var.get()
    m = thal_var.get()
    pred = {'A':[a],'B':[b],'C':[c],'D':[d], 'E':e, 'F':[f], 'G':[g], 'H':[h], 'I':[i], 'J':[j], 'K':[k], 'L':[l],'M':[m]}
    test_features = pd.DataFrame(pred)
    #use Random Forest's predict method to predict values
    prediction = rf.predict(test_features)
    if(prediction == 1):
        ans_var.set('Less Likely to have Heart Disease')
    else:
        ans_var.set('Likely to have Heart Disease')
    
master = Tk()

Label( master , text = 'DIAGNOSIS', font = ("Arial" , 15), fg = "blue", pady = 2).grid( row = 0 , column = 1 , columnspan = 3)
Label( master , text = 'Enter Your Age :' , font = ("Arial" , 10) , width = 25 ,  anchor = 'w' , pady = 10).grid( row = 2 , column = 1)
Label( master , text = '(in Years)' , font  = ("Arial" , 10) , anchor = 'w' , pady = 10).grid( row = 2 , column =3)
Label( master , text = ' ' , width = 2).grid( row = 2 , column = 4)
Label( master , text = ' ' , width = 2).grid( row = 2 , column = 0)
Label( master , text = 'Enter Your Gender :' , font = ("Arial", 10), width = 25 , anchor = 'w' , pady = 10).grid( row = 3 , column = 1)
Label( master , text = 'Select Chest Pain Type :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 4 , column = 1 , rowspan = 2)
Label( master , text = 'Resting Blood Pressure :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 6 , column = 1)
Label( master , text = 'Serum Cholestoral :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 7 , column = 1)
Label( master , text = '(in mm Hg)' , font=("Arial", 10), anchor = 'w' , pady =10).grid( row = 6 , column = 3)
Label( master , text = '(in Mg/dL)' , font=("Arial", 10), anchor = 'w' , pady =10).grid( row = 7 , column = 3)
Label( master , text = 'Fasting Blood Sugar > 120 Mg/dL :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 8 , column = 1)
Label( master , text = 'Resting Cardiographic Results :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 9 , column = 1 , rowspan= 2)
Label( master , text = 'Maximum Heart Rate Achieved :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 11 , column = 1)
Label( master , text = 'Exercise Induced Anigma :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 12 , column = 1)
Label( master , text = 'Oldpeak ST :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 13 , column = 1)
Label( master , text = 'Slope of Peak Exercise :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 14 , column = 1 , rowspan = 2)
Label( master , text = 'Vessels Coloured by Fluorosopy :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 16 , column = 1 , rowspan = 2)
Label( master , text = 'Thal :' , font=("Arial", 10), width = 25, anchor = 'w' , pady =10).grid( row = 18 , column = 1, rowspan = 2)

age_var = IntVar()
gender_var = IntVar()
cpt_var = IntVar()
rbp_var = IntVar()
sc_var = IntVar()
fbs_var = IntVar()
rer_var = IntVar()
mhr_var = IntVar()
eia_var = IntVar()
ops_var = DoubleVar()
sop_var = IntVar()
nmv_var = IntVar()
thal_var = IntVar()

e1 = Entry( master , textvariable = age_var , fg = "blue" , width = 20).grid( row = 2, column = 2)
g1 = Radiobutton( master , text = 'Male' , variable = gender_var , value = 1 , anchor = 'w' , width = 15).grid( row = 3 , column = 2)
g2 = Radiobutton( master , text = 'Female' , variable = gender_var , value = 0 , anchor = 'w' , width = 15).grid( row = 3 , column = 3)
c1 = Radiobutton( master , text = 'Typical Angina' , variable = cpt_var, value = 1 , anchor = 'w' , width = 15).grid( row = 4 , column = 2) 
c2 = Radiobutton( master , text = 'Atypical Angina' , variable = cpt_var, value = 2 , anchor = 'w' , width = 15).grid( row = 4 , column = 3)
c3 = Radiobutton( master , text = 'Non-Anginal Pain' , variable = cpt_var, value = 3 , anchor = 'w' , width = 15).grid( row = 5 , column = 2) 
c4 = Radiobutton( master , text = 'Asymptom' , variable = cpt_var, value = 4 , anchor = 'w' , width = 15).grid( row = 5 , column = 3)
e2 = Entry( master , textvariable = rbp_var , fg = "blue" , width = 20).grid( row = 6, column = 2)
e3 = Entry( master , textvariable = sc_var , fg = "blue" , width = 20).grid( row = 7, column = 2)
fbs1 = Radiobutton( master , text = 'Yes' , variable = fbs_var , value = 1 , anchor = 'w' , width = 15).grid( row = 8 , column = 2)
fbs2 = Radiobutton( master , text = 'No' , variable = fbs_var , value = 0 , anchor = 'w' , width = 15).grid( row = 8 , column = 3)
rer1 = Radiobutton( master , text = '0' , variable = rer_var , value = 0 , anchor = 'w' , width = 15).grid( row = 9 , column = 2)
rer2 = Radiobutton( master , text = '1' , variable = rer_var , value = 1 , anchor = 'w' , width = 15).grid( row = 9 , column = 3)
rer3 = Radiobutton( master , text = '2' , variable = rer_var , value = 2 , anchor = 'w' , width = 15).grid( row = 10 , column = 2)

mhr2 = Entry( master , textvariable = mhr_var , fg = "blue" , width = 20).grid( row = 11, column = 2)
eia1 = Radiobutton( master , text = 'Yes' , variable = eia_var , value = 1 , anchor = 'w' , width = 15).grid( row = 12 , column = 2)
eia2 = Radiobutton( master , text = 'No' , variable = eia_var , value = 0 , anchor = 'w' , width = 15).grid( row = 12 , column = 3)
ops1 = Entry( master , textvariable = ops_var , fg = "blue" , width = 20).grid( row = 13, column = 2)
sop1 = Radiobutton( master , text = 'Up Sloping' , variable = sop_var , value = 1 , anchor = 'w' , width = 15).grid( row = 14 , column = 2)
sop2 = Radiobutton( master , text = 'Flat' , variable = sop_var , value = 2 , anchor = 'w' , width = 15).grid( row = 14 , column = 3)
sop3 = Radiobutton( master , text = 'Down Sloping' , variable = sop_var , value = 3 , anchor = 'w' , width = 15).grid( row = 15 , column = 2)
nmv1 = Radiobutton( master , text = '0' , variable = nmv_var , value = 1 , anchor = 'w' , width = 15).grid( row = 16 , column = 2)
nmv2 = Radiobutton( master , text = '1' , variable = nmv_var , value = 2 , anchor = 'w' , width = 15).grid( row = 16 , column = 3)
nmv3 = Radiobutton( master , text = '2' , variable = nmv_var , value = 3 , anchor = 'w' , width = 15).grid( row = 17 , column = 2)
nmv4 = Radiobutton( master , text = '3' , variable = nmv_var , value = 4 , anchor = 'w' , width = 15).grid( row = 17 , column = 3)
thal1 = Radiobutton( master , text = 'Normal' , variable = thal_var , value = 3 , anchor = 'w' , width = 15).grid( row = 18 , column = 2)
thal2 = Radiobutton( master , text = 'Fixed Defect' , variable = thal_var , value = 6 , anchor = 'w' , width = 15).grid( row = 18 , column = 3)
thal3 = Radiobutton( master , text = 'Reversible Defect' , variable = thal_var , value = 7 , anchor = 'w' , width = 15).grid( row = 19 , column = 2)

button = Button( master , text = 'Show Result' , width = 20 , font=("Arial", 12), fg = "blue", command = calc).grid( row = 21 , column = 2)
Label( master , text = 'Query Result :' , width = 20 , font=("Arial", 12) , fg = "blue").grid( row = 22 , column = 1)
ans_var = StringVar()
Label( master , textvariable = ans_var , width = 40 , font=("Arial", 12) , fg = "Red").grid( row = 22 , column = 2 , columnspan = 2)
ans_var.set('ENTER DATA TO GET RESULT')

master.mainloop()
```
