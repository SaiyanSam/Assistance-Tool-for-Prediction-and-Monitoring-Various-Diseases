```python
from tkinter import *
import pandas as pd

prec = pd.read_excel('C://CAPSTONE//DataSets//Generalised//Precaution.xlsx')
info = pd.read_excel('C://CAPSTONE//DataSets//Generalised//Description.xlsx')
index = pd.read_csv('C://CAPSTONE//DataSets//Generalised//CSV//Indexing.csv')
index = index['Disease']

# symptoms
symps = ['nodal skin eruptions' , 'dischromic patches']
# severities symptoms wise
sever = [4 , 6]
# number of symptoms
num = 2

# denoting disease id
prediction = 16

master1 = Tk()

ans = StringVar()
desc = StringVar()
prec1 =StringVar()
prec2 =StringVar()
prec3 =StringVar()
prec4 =StringVar()

ans.set(index[prediction-1])
desc.set(info['Description'][prediction-1])
prec1.set(prec['Precaution_1'][prediction-1])
prec2.set(prec['Precaution_2'][prediction-1])
prec3.set(prec['Precaution_3'][prediction-1])
prec4.set(prec['Precaution_4'][prediction-1])

Label( master1 , text = 'RESULT', font=("Arial", 15), fg = "blue", pady = 5).grid( row = 0 , column = 1, columnspan = 2)
Label( master1 , text = 'Disease :' , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = 2 , column = 1)
Label( master1 , textvariable = ans , font=("Arial", 10), width = 40, pady = 10 , anchor ='w').grid( row = 2 , column = 2)
Label( master1 , text = '     ').grid( row = 3 , column = 4)
Label( master1 , text = '     ').grid( row = 2 , column = 0)

Label( master1 , text = 'Description :' , font=("Arial", 10), width = 15 , anchor ='w').grid( row = 3 , column = 1)
T1 = Text( master1 , height = 5, width = 40 , fg = "Blue")
T1.grid( row = 3 , column = 2)
scrollbar1 = Scrollbar(master1)
scrollbar1.grid( row = 3 , column = 4 , ipady = 15)
T1.config(yscrollcommand = scrollbar1.set) 
scrollbar1.config(command = T1.yview)
T1.insert(END, desc.get())
T1.config(state = DISABLED)
    
Label( master1 , text = 'Precaution 1 :' , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = 4 , column = 1)
Label( master1 , textvariable = prec1 , font=("Arial", 10), width = 40, pady = 10 , anchor ='w').grid( row = 4 , column = 2)

Label( master1 , text = 'Precaution 2 :' , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = 5 , column = 1)
Label( master1 , textvariable = prec2 , font=("Arial", 10), width = 40, pady = 10 , anchor ='w').grid( row = 5 , column = 2)

Label( master1 , text = 'Precaution 3 :' , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = 6 , column = 1)
Label( master1 , textvariable = prec3 , font=("Arial", 10), width = 40, pady = 10 , anchor ='w').grid( row = 6 , column = 2)

Label( master1 , text = 'Precaution 4 :' , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = 7 , column = 1)
Label( master1 , textvariable = prec4 , font=("Arial", 10), width = 40, pady = 10 , anchor ='w').grid( row = 7 , column = 2)
Label( master1 , text = 'Severity of Symtoms Added : ' , font=("Arial", 10), width = 30, pady = 10 , anchor ='w').grid( row = 8 , column = 1 , columnspan =2)

for x in range(num):
    s = DoubleVar()
    s = (sever[x]/7)*10
    col = StringVar()
    pvar = StringVar()
    p = StringVar()
    pvar = str(round(s,2))
    pvar = pvar+' out of 10'
    
    Label( master1 , text = symps[x] , font=("Arial", 10), width = 15, pady = 10 , anchor ='w').grid( row = (9+x) , column = 1)
    #decide color based on severity
    if(sever[x]>=6):
        col = 'Red'
    elif(sever[x]>=3):
        col = 'Orange'
    else:
        col = 'Green'
    Label(master1 , text = pvar, width = 40, bg = col).grid( row = (9+x) , column = 2)
    p.set(pvar)

master1.mainloop()
    
```
